### Step 1. VADER Seniment Scoring
We will use NLTK's SentimentIntensityAnalyzer to get the neg/neu/pos scores of the text.

This uses a "bag of words" approach:
Stop words are removed
each word is scored and combined to a total score.

```python
#nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('I am so happy!')
```

```python
sia.polarity_scores('This is the worst thing ever.')
```

```python
sia.polarity_scores(example)
```

### Run the polarity score on the entire dataset

```python
res = {}
for i, row in tqdm(data.iterrows(), total=len(data)):
    text = row['Text']
    myid = row['Id']
    polarity = sia.polarity_scores(text)
    if  polarity['compound'] >= 0.05 :
        res[myid] = "Positive"
    
    elif  polarity['compound'] <= - 0.05 :
        res[myid] ="Negative"
    
    else :
        res[myid] ="Neutral"# Now we have sentiment score and metadata
```

```python
import pandas as pd
data.head()
```

```python
res_df = pd.DataFrame(res.items())
res_df.rename(columns={0: 'Id' , 1:'Feedback'}, inplace =True)
res_df
```

```python
vaders = res_df.merge(data, how='inner')
vaders
```

```python
final_data = vaders[ ['Id' , 'Summary', 'Feedback'] ]
final_data
```
